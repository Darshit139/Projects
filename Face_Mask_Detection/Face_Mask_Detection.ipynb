```python
import pandas as pd
import numpy as np 
import os
from keras.preprocessing import image
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import VGG16
from keras import Sequential,Model
from keras.layers import Dense
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
path = 'C:\\Users\\DARSHIT\\Desktop\\Data Science and AI\\Deep Learning\\Projects\\Face_Mask_Detection\\train'
```

```python
#os.listdir(path)
```

```python
x = []
y = []

for i in os.listdir(path):
    img = cv2.imread(os.path.join(path,i))
    img = cv2.resize(img,(224,224))
    img = img/255.
    #print(img.shape)
    
    if 'with-mask' in i:
        label = 1
    else:
        label = 0
    x.append(img)  
    y.append(label)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
vgg = VGG16()
```

```python
vgg.summary()
```

```python
vgg.trainable = False
```

```python
model = Sequential()
```

```python
for i in vgg.layers[:-1]:
    model.add(i)
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_train.shape
```

```python
model.fit(x_train,y_train,epochs=5 ,validation_data = (x_test,y_test),batch_size=32)
```

```python
model.save('m1')
```

```python
cap = cv2.VideoCapture(0)
```

```python
def fun(img):
    y_pred = model.predict(img.reshape(1,224,224,3))
    return y_pred
```

```python
def fun2(img,text,pos,bg_color):
    
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    
    end_x = pos[0] + text_size[0][0] + 2
    end_y = pos[1] + text_size[0][1] - 2
    
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)
    
```

```python
while True:
    ret, frame = cap.read()
    
    # call detection method
    img = cv2.resize(frame,(224,224))
    y_pred = fun(img)
    print(y_pred)
    y_pred = np.argmax(y_pred)
    
    coods =fun3(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    
    for x,y,w,h in coods:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    
    if y_pred == 0:
        fun2(frame,'No Face Mask',(10,10),(0,255,0))
    else:     
        fun2(frame,'Face Mask',(10,10),(0,0,255))

    print(y_pred) 
    cv2.imshow('window',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()        
```

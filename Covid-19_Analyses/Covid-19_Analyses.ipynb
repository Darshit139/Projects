# Q.1 Which country/region have most covid 19 death count and how it treands around other top 15 countries in same respective.

```python
sumd=data2.groupby("Country/Region").sum()#.sort_values(by='Active',ascending=False)
sumd
```

```python
sumd.drop(['Lat', 'Long'],axis=1,inplace=True)
```

```python
sumd
```

```python
deaths=sumd.Deaths.sort_values(ascending=False) # Focusing on death count of each country
deaths
```

```python
# visualising the top 15 data in deaths variable.
deaths.head(15).plot(figsize=(7,7),kind="bar",grid=True,xlabel="Country",ylabel="Death Count",
                     title="Top 15 Countries for Covid 19 DEATH COUNT")
plt.show()
```
